```python
## Matching ip address 
def having_ip_address(url):
    match = re.search(
        '(([01]?\\d\\d?|2[0-4]\\d|25[0-5])\\.([01]?\\d\\d?|2[0-4]\\d|25[0-5])\\.([01]?\\d\\d?|2[0-4]\\d|25[0-5])\\.'
        '([01]?\\d\\d?|2[0-4]\\d|25[0-5])\\/)|'  # IPv4
        '((0x[0-9a-fA-F]{1,2})\\.(0x[0-9a-fA-F]{1,2})\\.(0x[0-9a-fA-F]{1,2})\\.(0x[0-9a-fA-F]{1,2})\\/)' # IPv4 in hexadecimal
        '(?:[a-fA-F0-9]{1,4}:){7}[a-fA-F0-9]{1,4}', url)  # Ipv6
    if match:
        #print ("1")
        return 1
    else:
        #print ('o')
        return 0
```

```python
final_df["use_of_ip"]=final_df["url"].apply(lambda i: having_ip_address(i)) ## inseting a coloumb
final_df["use_of_ip"].unique()
```

```python
# If the hostname extracted from a URL is found somewhere within the URL itself, it typically indicates that the URL structure might be abnormal or unconventional. Whether this is considered "abnormal" or "anomalous" depends on the context and the specific criteria you're using to define abnormality.
```

```python
def check_abnormality(url):
    hostname=urlparse(url).hostname ## it returns the host name if it exist otherwise return nothing
    hostname=str(hostname)
    match=re.search(hostname,url)
    if match : ## it give 
        return 0 ## why zero study prev cell
    else:
        return 1
    
```

```python
final_df["abnormal_url"]=final_df["url"].apply(lambda i: check_abnormality(i))
```

```python
## In the context of a URL (Uniform Resource Locator), netloc refers to the network location part of the URL. It typically includes the domain name (or IP address) and may also include a port number. The netloc is an essential part of a URL because it specifies the location of the resource on the internet.
```

```python
def hostname_length(url):
    return len(urlparse(url).netloc)
final_df['hostname_length'] = final_df['url'].apply(lambda i: hostname_length(i))
```

```python
def suspicious_words(url):
    match = re.search('PayPal|login|signin|bank|account|update|free|lucky|service|bonus|ebayisapi|webscr',
                      url)
    if match:
        return 1
    else:
        return 0
final_df['sus_url'] = final_df['url'].apply(lambda i: suspicious_words(i))
```

```python
def digit_count(url):
    digits = 0
    for i in url:
        if i.isnumeric():
            digits = digits + 1
    return digits
final_df['count-digits']= final_df['url'].apply(lambda i: digit_count(i))
```

```python
def letter_count(url):
    letters = 0
    for i in url:
        if i.isalpha():
            letters = letters + 1
    return letters
final_df['count-letters']= final_df['url'].apply(lambda i: letter_count(i))
```

```python
def fd_length(url):
    urlpath= urlparse(url).path
    try:  
        return len(urlpath.split('/')[1])
    except: ## if output is None 
        return 0

final_df['fd_length'] = final_df['url'].apply(lambda i: fd_length(i))
```
